```python
import pandas as pd
import os
```

```python
# First of all, get the path of data
PATH_IN = "./data/"
file = os.path.join(PATH_IN, "interactive_data.csv")
df = pd.read_csv(file, index_col=0)
# df head to check the data
df.head(7)
```

```python
# filtered none selected bc no info so they can be anything,  axis = 1 to iterate filter on lines (0 column)
df_filtered = df.loc[~(df == "None selected").any(axis=1)]
df_filtered.head(7)
```

```python
total_deaths = df_filtered["Deaths"].sum()

deaths_by_suicide = df_filtered[df_filtered["Intent"] == "Suicide"]["Deaths"].sum()
print(deaths_by_suicide / total_deaths * 100, "% of gun deaths are suicides")

male_deaths_by_suicide = df_filtered[
    (df_filtered["Intent"] == "Suicide") & (df_filtered["Gender"] == "Male")
]["Deaths"].sum()
print(
    male_deaths_by_suicide / deaths_by_suicide * 100, "% of male deaths in gun suicides"
)

homicide_deaths = df_filtered[df_filtered["Intent"] == "Homicide"]["Deaths"].sum()
print(homicide_deaths / total_deaths * 100, "% of homicide of gun deaths")

homicide_male_15_34_deaths = df_filtered[
    (df_filtered["Gender"] == "Male")
    & (df_filtered["Age"] == "15 - 34")
    & (df_filtered["Intent"] == "Homicide")
]["Deaths"].sum()
homicide_male_15_34_deaths_black = df_filtered[
    (df_filtered["Gender"] == "Male")
    & (df_filtered["Age"] == "15 - 34")
    & (df_filtered["Race"] == "Black")
    & (df_filtered["Intent"] == "Homicide")
]["Deaths"].sum()
print(
    homicide_male_15_34_deaths_black / homicide_male_15_34_deaths * 100,
    "%  are black in male 15 - 34 homicide deaths",
)

women_homicide_deaths = df_filtered[
    (df_filtered["Gender"] == "Female") & (df_filtered["Intent"] == "Homicide")
]["Deaths"].sum()
print(
    women_homicide_deaths / homicide_deaths * 100, "% of women in all homicide deaths"
)
```
